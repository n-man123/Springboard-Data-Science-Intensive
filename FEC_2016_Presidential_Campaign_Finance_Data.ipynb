```python
months = {'JAN' : 1, 'FEB' : 2, 'MAR' : 3, 'APR' : 4, 'MAY' : 5, 'JUN' : 6,
          'JUL' : 7, 'AUG' : 8, 'SEP' : 9, 'OCT': 10, 'NOV': 11, 'DEC' : 12}
```

```python
from datetime import datetime
```

```python
def convert_date(val):
    d, m, y = val.split('-')
    m = months[m]
    return datetime(int('20' + y), m, int(d))
```

```python
df.contb_receipt_dt.head(10)
```

```python
df['contb_receipt_dt'] = df.contb_receipt_dt.map(convert_date)
```

```python
df.contb_receipt_dt[0]
```

```python
df.groupby(['cand_nm', 'contb_receipt_dt'])['contb_receipt_amt'].sum()
```

```python
by_date = df.groupby(['cand_nm', 'contb_receipt_dt'])['contb_receipt_amt'].sum()
```

```python
type(by_date)
```

```python
by_date.unstack('cand_nm')
```

```python
new_df = by_date.unstack('cand_nm').cumsum()#.plot().legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
filtered_df = new_df.ix['2015-01-01':'2016-08-01']
```
